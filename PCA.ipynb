```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris() # loading the iris dataset.
X = iris["data"]
y = iris["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # splitting into train and test.
```

```python
x2d = PCA(n_components=2) # reducing components from 4 to 2.
x2d.fit_transform(X_train)
```
