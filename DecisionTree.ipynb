```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
iris = load_iris() # loading the irism dataset.
```

```python
X = iris.data[:, 2:] # only using 2 out of 4 coulumns.
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2) #decision tree classifier with depth 2.
tree_clf.fit(X, y)
```

```python
text_representation = tree.export_text(tree_clf) # displaying the tree in text form.
print(text_representation)
```

```python
fig = plt.figure(figsize=(15,15)) #displaying the tree in visual form.
_ = tree.plot_tree(tree_clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
import numpy as np
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()
```

```python
tree_reg = DecisionTreeRegressor(max_depth=2) # using Decision tree to solve linear regression problem.
tree_reg.fit(X, y)
```

```python
fig = plt.figure(figsize=(10, 10)) # predicting the values.
graph = tree.plot_tree(tree_reg,
                       filled=True
                      )
```
